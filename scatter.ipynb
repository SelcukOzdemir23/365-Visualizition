```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv("scatter_data.csv")
```

```python
data
```

```python
plt.figure(figsize=(12,8))
scatter = plt.scatter(
    data["Area (ft.)"],
    data["Price"],
    alpha=0.6,
    c = data["Building Type"],
    cmap = 'viridis'
)
plt.legend(*scatter.legend_elements(),
          loc="upper left",
          title="Building Type"

)
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(data=data,x=data["Price"],
                y =data["Area (ft.)"],
                hue=data["Building Type"],
                palette=["black","darkblue","purple","pink","white"],
                s=100
)
```
