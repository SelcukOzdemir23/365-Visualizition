```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv("histogram_data.csv")
```

```python
df
```

```python
sns.set_style("dark")
plt.figure(figsize=(8,6))
plt.hist(
    df["Price"],
    bins=8,
    color="#108A99"
    )

plt.title("Deneme")
plt.xlabel("dene")
plt.ylabel("Price in")
sns.despine()
plt.show()
```
