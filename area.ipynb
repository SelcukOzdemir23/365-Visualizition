```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv("stacked_area_chart_data.csv")
```

```python
df
```

```python
colors =["#011638","#7e2987","#ef2026"]
sns.set_style("white")
labels = ["Diesel","Petrol","Gas"]
plt.figure(figsize=(12,6))
plt.stackplot(
    df["Year"],
    df["Diesel"],
    df["Petrol"],
    df["Gas"],
    colors=colors,
    edgecolor="none"
)

plt.xticks(df["Year"],rotation=45)
plt.legend(labels)
plt.ylabel("Number Of Cars")
plt.title("Popularity of Engine Fuel Types(1982-2016)")
sns.despine()
plt.show()
```
