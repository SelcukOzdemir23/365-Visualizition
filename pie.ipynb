```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv("pie_chart_data.csv")
```

```python
data
```

```python
sns.set_palette("colorblind")
```

```python
plt.figure(figsize=(10,8))
```

```python
plt.pie(data["Number of Cars"],labels = data["Engine Fuel Type"].values,
autopct="%.2f%%",textprops={"size":"medium","fontweight":"bold","rotation":30,"color":"w"})
plt.legend()
plt.title("Cars by engine Fuel Type",fontsize=18)
```
