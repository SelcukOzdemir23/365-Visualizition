```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv("line_chart_data.csv")
```

```python
df
```

```python
df["Date"]= pd.to_datetime(df["Date"])
```

```python
df
```

```python
labels = ["S&P 500","FTSE 100"]
plt.figure(figsize=(20,8))
plt.plot(df["Date"],df["GSPC500"])
plt.plot(df["Date"],df["FTSE100"])
plt.legend(labels)
plt.show()
```

```python
df2 = df[(df["Date"]>="2008-07-01") & (df["Date"]<="2008-12-31")]
```
